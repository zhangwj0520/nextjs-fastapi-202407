
# Bar color demo

This is an example showing how to control bar color and legend entries
using the *color* and *label* parameters of `~matplotlib.pyplot.bar`.
Note that labels with a preceding underscore won't show up in the legend.


```python
import matplotlib.pyplot as plt


fig, ax = plt.subplots()

# 1.背景颜色
ax.patch.set_color('#175C97')
fig.patch.set_facecolor('#175C97')
# 2.文字颜色
# 标题内容,颜色,位置
plt.title("title", color="white",loc='left')
# X轴文字
plt.xlabel("X value", color="white", loc='right')
# Y轴文字
plt.ylabel("fruit supply", color="white")
# XY轴颜色
plt.xticks(color="white")
plt.yticks(color="white")
plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors='white')
# 3 折线图,柱状图配色方案,后续改成动态
colors = ['#4992ff',
  '#7cffb2',
  '#fddd60',
  '#ff6e76',
  '#58d9f9',
  '#05c091',
  '#ff8a45',
  '#8d48e3',
  '#dd79ff']




fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']


ax.bar(fruits, counts, label=bar_labels, color=colors)

# 4 图例
ax.legend(labelcolor='white',facecolor='#175C97')
ax.spines['bottom'].set_color('white')
ax.spines['top'].set_color('#175C97')
ax.spines['right'].set_color('#175C97')
ax.spines['left'].set_color('white')



# 3.背景网格 ,存在覆盖问题,暂时不设置了
# ax.grid(True, which='both', linestyle='-', linewidth=0.5, color='#817f91', alpha=0.2, zorder=-10)




plt.show()
```
